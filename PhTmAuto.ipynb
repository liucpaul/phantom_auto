```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

df = pd.read_csv("c:/users/paul/downloads/phantom auto/test.csv")
df.head()
```

```python
df['Velocity'].dtype
```

```python
# need to convert speed value to numeric
vel = pd.to_numeric(df['Velocity'], errors='coerce')
rsrp = df['RSRP_']
sinr = df['SINR_']
print(vel.shape, rsrp.shape, sinr.shape, vel.dtype, rsrp.dtype, sinr.dtype)
```

```python
# correlation check
# speed has little correlation to signal strenth or quality
print(vel.corr(rsrp), vel.corr(sinr), rsrp.corr(sinr))
```

```python
plt.scatter(sinr, rsrp)
plt.show()
```

```python
plt.scatter(vel, sinr)
plt.show()
```

```python
plt.scatter(vel, rsrp)
plt.show()
```
